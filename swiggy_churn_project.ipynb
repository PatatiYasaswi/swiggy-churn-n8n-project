```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

df = pd.DataFrame({
    "customer_id": range(10001, 10001+n),
    "age": np.random.randint(18, 60, n),
    "city": np.random.choice(
        ["Bangalore", "Hyderabad", "Chennai", "Mumbai"],
        n
    ),
    "orders_last_30_days": np.random.randint(0, 20, n),
    "avg_order_value": np.random.randint(100, 1500, n),
    "days_since_last_order": np.random.randint(0, 90, n),
    "coupon_used_last_month": np.random.choice([0,1], n),
    "customer_support_calls": np.random.randint(0,10,n)
})

# Churn Logic
df["churn"] = np.where(
    (
        (df["days_since_last_order"] > 45)
        &
        (df["orders_last_30_days"] < 5)
    ),
    1,
    0
)

df.to_csv("swiggy_customer_data.csv", index=False)

print(df.head())
print(df.shape)
```

# Model traning

```python
from sklearn.model_selection import train_test_split
```

```python
df_clean = pd.read_csv("swiggy_customer_data_cleaned.csv")
```

```python
features = [
    "age",
    "orders_last_30_days",
    "avg_order_value",
    "days_since_last_order",
    "coupon_used_last_month",
    "customer_support_calls"
]
# features to predict churn
```

```python
X = df_clean[features]
y = df_clean["churn"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42
)
```

```python
print("Training set size:", X_train.shape[0])
print("Testing set size:", X_test.shape[0])
```

```python
print("\nTraining churn distribution:")
print(y_train.value_counts())
```

```python
print("\nTest churn distribution:")
print(y_test.value_counts())
```

# Classificatio model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
print("Model Accuracy:", accuracy)
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)
```

```python
df_clean["predicted_churn"] = model.predict(X)
```

```python
df_clean["customer_segment"] = df_clean["predicted_churn"].map({
    1: "Inactive (Churned)",
    0: "Regular (Non-churned)"
})
```

```python
df_clean["marketing_action"] = df_clean["customer_segment"].map({
    "Inactive (Churned)": "Send coupon on previous purchase category",
    "Regular (Non-churned)": "Send thank-you message and loyalty reward"
})
```

```python
print("Customer segment summary:")
print(df_clean["customer_segment"].value_counts())

print("\nMarketing action summary:")
print(df_clean["marketing_action"].value_counts())
```

```python
print("\nExample inactive customers:")
inactive = df_clean[df_clean["customer_segment"] == "Inactive (Churned)"].head(5)
print(inactive[["customer_id", "days_since_last_order", "orders_last_30_days", "customer_segment", "marketing_action"]])

print("\nExample regular customers:")
regular = df_clean[df_clean["customer_segment"] == "Regular (Non-churned)"].head(5)
print(regular[["customer_id", "days_since_last_order", "orders_last_30_days", "customer_segment", "marketing_action"]])

# Save the final table with segments and actions
df_clean.to_csv("swiggy_customer_segments_with_actions.csv", index=False)
print("\nFinal CSV saved: swiggy_customer_segments_with_actions.csv")
```
